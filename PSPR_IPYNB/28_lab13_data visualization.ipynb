```python
def f(sex, age):
    return np.sin(np.sqrt(sex ** 2 + age ** 2))

sex = np.linspace(-6, 6, 30)
age= np.linspace(-6, 6, 30)

sex, age = np.meshgrid(x, y)
chol = f(sex, age)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(sex, age, chol, 50, cmap='binary')
ax.set_xlabel('sex')
ax.set_ylabel('age')
ax.set_zlabel('chol');
```
