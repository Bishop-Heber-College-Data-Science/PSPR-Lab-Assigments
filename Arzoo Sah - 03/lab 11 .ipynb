1. Retrieve data from web page using URLLIB and print the frequency of words from that page.

```python
import urllib.request
counts = dict()
fhand = urllib.request.urlopen('http://www.py4inf.com/code/romeo.txt')
for line in fhand:
    words = line.split()
    for word in words:
        counts[word] = counts.get(word,0) + 1
print(counts)
```

4 Create a JSON file for the following Students Marks and print the number of students
and their names and marks.
ID: DS01:  Name: rex
           Semester1: 80, 55
           Semester2: 50, 70, 82
ID: DS02:  Name: peter
Semester1: 92, 75

```python
import json
data = '''
[
{ "id" : "DS01",
  "Name" : "rex",
  "Semester1" : "80,55",
  "Semester2" : "50,70"
},
{ "id" : "DS02",
  "Name" : "peter", 
  "Semester1" : "92,75",
  "Semester2" : "64,87"
}
]'''
info = json.loads(data)
print('User count:', len(info))
for item in info:
    print('Name', item['Name'])
    print('Id', item['id'])
    print('Semester1', item['Semester1'])
    print('Semester2', item['Semester2'])
```
